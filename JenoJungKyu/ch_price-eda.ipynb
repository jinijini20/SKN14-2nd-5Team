```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Windows용 한글 폰트 지정
plt.rcParams['font.family'] = 'Malgun Gothic'

# 마이너스 기호 정상 표시 (음수 깨짐 방지)

plt.rcParams['axes.unicode_minus'] = False
```

```python
customers_df = pd.read_csv('data/olist_customers_dataset.csv')
sellers_df = pd.read_csv('data/olist_sellers_dataset.csv')
orders_df = pd.read_csv('data/olist_orders_dataset.csv')
products_df = pd.read_csv('data/olist_products_dataset.csv')
geolocation_df = pd.read_csv('data/olist_geolocation_dataset.csv')
order_items_df = pd.read_csv('data/olist_order_items_dataset.csv')
order_reviews_df = pd.read_csv('data/olist_order_reviews_dataset.csv')
categories_df = pd.read_csv('data/product_category_name_translation.csv')
```

```python
#  customer_unique_id 결합
orders_df = orders_df.merge(customers_df[['customer_id', 'customer_unique_id']], on='customer_id', how='left')

#  고객별 주문 수 계산
order_counts = orders_df.groupby('customer_unique_id')['order_id'].nunique().reset_index()
order_counts.columns = ['customer_unique_id', 'order_count']

#  2회 이상 구매한 고객 필터링
repeat_customers = order_counts[order_counts['order_count'] >= 2]
print(f"재구매 고객 수: {len(repeat_customers)}명")

#  재구매 고객의 주문만 필터링
repeat_orders = orders_df[orders_df['customer_unique_id'].isin(repeat_customers['customer_unique_id'])]
print(f"재구매 고객 주문 수: {len(repeat_orders)}건")

#  주문별 가격 정보 결합
merged = pd.merge(repeat_orders, order_items_df, on='order_id', how='inner')

#  고객별 재구매 횟수, 평균 가격 계산
cust_summary = merged.groupby('customer_unique_id').agg(
    purchase_count=('order_id', 'nunique'),
    avg_price=('price', 'mean')
).reset_index()

#  가격 구간 나누기
cust_summary['price_bin'] = pd.qcut(cust_summary['avg_price'], 10, duplicates='drop')

#  가격 구간별 평균 재구매 횟수
avg_purchase_by_price_bin = cust_summary.groupby('price_bin', observed=True)['purchase_count'].mean()

# . 시각화
plt.figure(figsize=(10,6))
avg_purchase_by_price_bin.plot(marker='o')
plt.title('가격대별 평균 재구매 횟수 (재구매 고객 기준)')
plt.xlabel('가격 구간 (BRL)')
plt.ylabel('평균 재구매 횟수')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
 # 고객별 평균 가격 및 구매 수 계산
cust_summary = merged.groupby('customer_unique_id').agg({
    'price': 'mean',
    'order_id': 'nunique'
}).rename(columns={'order_id': 'purchase_count'}).reset_index()

# 가격을 10구간으로 나눔
cust_summary['price_bin'] = pd.qcut(cust_summary['price'], 10)

# 가격 구간별 평균 구매 수 시각화
avg_purchase_by_price_bin = cust_summary.groupby('price_bin')['purchase_count'].mean()

plt.figure(figsize=(12,6))
avg_purchase_by_price_bin.plot(kind='bar')
plt.title('가격 구간별 평균 재구매 수')
plt.ylabel('평균 재구매 횟수')
plt.xlabel('가격 구간')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
# 거의 골고르 2번씩 삼 ㄷㄷ
```

# 재구매 고객 정의 및 주문 필터링

```python
# customer_unique_id 결합
orders_df = orders_df.merge(customers_df[['customer_id', 'customer_unique_id']], on='customer_id', how='left')

# 고객별 주문 수 계산
order_counts = orders_df.groupby('customer_unique_id')['order_id'].nunique().reset_index()
# 언유니크하면 컬럼이 바뀐값이 반환되니까 리셋인덱스를 해주는겨

order_counts.columns = ['customer_unique_id', 'order_count']

# 2회 이상 구매 고객 추출
repeat_customers = order_counts[order_counts['order_count'] >= 2]

# 재구매 고객 주문 필터링
repeat_orders = orders_df[orders_df['customer_unique_id'].isin(repeat_customers['customer_unique_id'])]
```

```python
repeat_orders
```

```python
# 날짜 타입으로 변환
repeat_orders['order_purchase_timestamp'] = pd.to_datetime(repeat_orders['order_purchase_timestamp'])

# 고객별 마지막 구매일 계산
last_purchase = repeat_orders.groupby('customer_unique_id')['order_purchase_timestamp'].max().reset_index()
last_purchase.columns = ['customer_unique_id', 'last_purchase_date']
```

```python
last_purchase
```

```python
# 전체 주문 중 가장 마지막 날짜 기준
reference_date = repeat_orders['order_purchase_timestamp'].max()

# 마지막 구매일로부터 경과일 계산
last_purchase['days_since_last_purchase'] = (reference_date - last_purchase['last_purchase_date']).dt.days

# 이탈 여부 정의: 90일 이상 경과 → 이탈
last_purchase['churn'] = last_purchase['days_since_last_purchase'] > 90

# 결과 확인
print("재구매 고객 수:", len(last_purchase))
print("이탈 고객 수:", last_purchase['churn'].sum())
last_purchase
```

```python
# 상품 카테고리 정보 결합
orders_with_items = pd.merge(repeat_orders[['order_id', 'customer_unique_id']],
                             order_items_df[['order_id', 'product_id']],
                             on='order_id', how='left')

orders_with_product = pd.merge(orders_with_items, products_df[['product_id', 'product_category_name']],
                               on='product_id', how='left')

# 고객별 대표 카테고리 선정 (가장 많이 구매한 것)
top_category_per_customer = orders_with_product.groupby(['customer_unique_id', 'product_category_name']).size().reset_index(name='count')
top_category = top_category_per_customer.sort_values(['customer_unique_id', 'count'], ascending=[True, False]).drop_duplicates('customer_unique_id')

# 이탈 정보 결합
churn_category_df = pd.merge(last_purchase, top_category[['customer_unique_id', 'product_category_name']], on='customer_unique_id', how='left')

# 카테고리별 이탈률
churn_by_category = churn_category_df.groupby('product_category_name')['churn'].mean().reset_index().sort_values(by='churn', ascending=False)

# 상위 10개 시각화
plt.figure(figsize=(12,6))
sns.barplot(data=churn_by_category.head(10), x='product_category_name', y='churn', palette='coolwarm')
plt.title('상품 카테고리별 이탈률 (상위 10개)')
plt.ylabel('이탈률')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 2. orders_df에 customer_unique_id 병합
orders_df = orders_df.merge(customers_df[['customer_id', 'customer_unique_id']], on='customer_id', how='left')

# 3. 고객별 주문 수 계산 (재구매 고객 찾기)
order_counts = orders_df.groupby('customer_unique_id')['order_id'].nunique().reset_index()
order_counts.columns = ['customer_unique_id', 'order_count']
repeat_customers = order_counts[order_counts['order_count'] >= 2]

# 4. 재구매 고객 주문만 필터링
repeat_orders = orders_df[orders_df['customer_unique_id'].isin(repeat_customers['customer_unique_id'])]

# 5. 중복 컬럼 제거 (merge 과정에서 생긴 중복 컬럼 문제 방지)
repeat_orders = repeat_orders.loc[:, ~repeat_orders.columns.duplicated()]

# 6. 날짜 타입 변환
repeat_orders['order_purchase_timestamp'] = pd.to_datetime(repeat_orders['order_purchase_timestamp'])

# 7. 고객별 마지막 구매일 계산
last_purchase = repeat_orders.groupby('customer_unique_id')['order_purchase_timestamp'].max().reset_index()
last_purchase.columns = ['customer_unique_id', 'last_purchase_date']

# 8. 전체 주문 중 마지막 날짜 기준
reference_date = repeat_orders['order_purchase_timestamp'].max()

# 9. 마지막 구매일 기준 경과일 계산
last_purchase['days_since_last_purchase'] = (reference_date - last_purchase['last_purchase_date']).dt.days

# 10. churn 컬럼 정의 (90일 이상 경과 시 이탈로 간주)
last_purchase['churn'] = last_purchase['days_since_last_purchase'] > 90

# 11. 주문별 가격 정보 병합
repeat_orders = repeat_orders.merge(order_items_df[['order_id', 'price']], on='order_id', how='left')

# 12. 고객별 총 지출, 평균 주문 금액, 주문 횟수, 첫 구매, 마지막 구매 계산
customer_value = repeat_orders.groupby('customer_unique_id').agg({
    'price': ['sum', 'mean'],
    'order_purchase_timestamp': ['min', 'max', 'count']
}).reset_index()

customer_value.columns = ['customer_unique_id', 'total_spent', 'avg_order_value',
                          'first_purchase', 'last_purchase', 'order_count']

# 13. 평균 구매 간격 계산
customer_value['purchase_frequency'] = (
    (customer_value['last_purchase'] - customer_value['first_purchase']).dt.days /
    (customer_value['order_count'] - 1)
)
customer_value['purchase_frequency'].fillna(0, inplace=True)

# 14. churn 정보 병합
customer_features = customer_value.merge(last_purchase[['customer_unique_id', 'churn']], on='customer_unique_id', how='left')

# 15. 상관계수 히트맵 그리기
corr_cols = ['total_spent', 'avg_order_value', 'order_count', 'purchase_frequency', 'churn']
corr_matrix = customer_features[corr_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('고객 특성 상관계수 히트맵')
plt.show()
```

```python
# 상관계수 0.86
# avg_order_value(평균 주문 금액)가 높은 고객일수록 전체 지출 금액도 높다는 경향이 매우 강하다.
# 즉, 비싼 걸 사는 고객은 많이도 산다는 경향이 있는 것처럼 보임.
# 확인결과 그냥 주문건수랑 비례함 평균이라 의미가 무색
```

```python
#  상관계수 0.3의 의미
# 0.3 정도는 통계적으로 약한 양의 상관관계
# 즉, 주문 횟수가 많다고 해서 반드시 지출 금액이 높은 것은 아니다.
# 이것도 의미가...왜냐면 평균이긴하지만 저가형고객데이터가 거의 대부분이므로..표본이 고르지않음
```
